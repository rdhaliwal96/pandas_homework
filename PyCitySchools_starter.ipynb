```python
# Dependencies and Setup
import os
import pandas as pd
import numpy as np


# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

```python
#printed each function so i could cofirm the data i was pulling was accurate

total_schools = school_data_complete["School ID"].count()
total_schools

total_students = school_data_complete["Student ID"].count()
total_students

total_budget = school_data_complete["budget"].sum()
total_budget

average_math_score = school_data_complete['math_score'].mean()
average_math_score

average_reading_score = school_data_complete['reading_score'].mean()
average_reading_score

passing_math_score = (school_data_complete['math_score']>=70).sum()/total_students
passing_math_score*100

passing_reading_score = (school_data_complete['reading_score']>=70).sum()/total_students
passing_reading_score*100

passing_overall_score = (passing_math_score+passing_reading_score)/2
passing_overall_score*100

```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#created new variable types

school_name = school_data_complete["school_name"]
school_type = school_data_complete["type"]
students = school_data_complete["size"]
budget = school_data_complete["budget"]
math_average = school_data_complete.groupby(["school_name"])["math_score"].mean()
reading_average = school_data_complete.groupby(["school_name"])["reading_score"].mean()
math_passing = school_data_complete[school_data_complete.math_score>=70].groupby('school_name')['math_score'].count()/school_data_complete.groupby(["school_name"])["Student ID"].count()
math_passing = math_passing *100
reading_passing = school_data_complete[school_data_complete.reading_score>=70].groupby('school_name')['reading_score'].count()/school_data_complete.groupby(["school_name"])["Student ID"].count()
reading_passing = reading_passing*100
overall_passing = (math_passing + reading_passing)/2

school_overview1_df = pd.DataFrame({
    "School Name":school_name,
    "School Type":school_type,
    "Total Students":students,
    "Total Budget":total_budget,
    "Budget Per School":budget
})
school_overview2_df = pd.DataFrame({
    "School Name":school_name,
    "Avg Math Score" :math_average,
    "Avg Reading Score":reading_average,
    "Percentage Passing Math":math_passing,
    "Percentage Passing Reading":reading_passing,
    "Percentage Passing Overall":overall_passing
})

school_overview = school_overview1_df.merge(school_overview2_df, on="School Name" ,how='right')
school_overview.head()

```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
budget_per_student = budget/students
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]




school_overview["Spending Ranges"] = pd.cut(budget_per_student,spending_bins,labels=group_names)

spending_math = school_overview.groupby("Spending Ranges").mean()["Avg Math Score"]
spending_reading = school_overview.groupby("Spending Ranges").mean()["Avg Reading Score"]
spending_math_pass = school_overview.groupby("Spending Ranges").mean()["Percentage Passing Math"]
spending_read_pass = school_overview.groupby("Spending Ranges").mean()["Percentage Passing Reading"]
spending_overall_pass = (spendingmath+spendingreading)/2

spending_summary = pd.DataFrame({
    "Average Math Score":spending_math,
    "Average Reading Score":spending_reading,
    "Passing Math Score":spending_math_pass,
    "Passing Reading Score":spending_read_pass,
    "Passing Overall Score":spending_overall_pass
})
spending_summary
```
